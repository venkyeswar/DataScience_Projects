```python
df.rename({"Area_of_the_house(excluding_basement)":"area_of_house_excluding_basement"},axis=1,inplace=True)
```

```python
import statsmodels.formula.api as smf
columns=df.columns
X = columns.drop("Price")
VIF = {}
for column in X:
    inputs = X.drop(column)
    inputs= "+".join(inputs)
    r2_score = smf.ols(f"{column}~{inputs}",data=df).fit().rsquared
    vif =1/(1-r2_score)
    VIF[column] = vif

```

```python
# Building Linear Regression Model
import statsmodels.formula.api as smf
columns = X.columns
input = "+".join(columns)
linear_model = smf.ols(f"Price~{input}",data=df).fit()
```

```python
linear_model.summary()
```

```python
# Testing the error mean
np.round(linear_model.resid.mean())
```

```python
# Plotting qq plot
import statsmodels.api as sm
sm.qqplot(linear_model.resid,line="q")
plt.show()
```

```python
# Homoscedasticity Test
def get_stand(x):
    return (x-x.mean())/x.std()

plt.scatter(get_stand(linear_model.fittedvalues),get_stand(linear_model.resid))
```

```python
# Cooks Distance
model_influence = linear_model.get_influence()
c,_ = model_influence.cooks_distance
plt.stem(np.arange(len(df)),np.round(c,3))
plt.show()
```

```python
# R2_Score Validation
from sklearn.metrics import r2_score
print(np.round(r2_score(Y,linear_model.predict(X))*100,2),"%")
```
